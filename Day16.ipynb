```python
import pandas as np
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_data=sns.load_dataset('iris')
iris_data
```

```python
sns.boxplot(x=iris_data['sepal_width'],palette=['violet','red','lightgreen'],y=iris_data['species'])
plt.show()
```

```python
import matplotlib.pyplot as plt
sns.boxplot(x=iris_data['sepal_width'],color='red')
plt.show()
```

```python
df=sns.load_dataset('taxis')
sns.scatterplot(x=df['fare'],y=df['total'],palette=['yellow','red'])
plt.show()
```

```python
df=sns.load_dataset('taxis')
sns.scatterplot(x=df['fare'],y=df['total'],palette=['darkblue','red'],hue=df['payment'])
plt.show()
```

```python
sns.stripplot(x=df['fare'],y=df['payment'])
```

```python
sns.stripplot(x=df['fare'],y=df['payment'],palette=['yellow','red'],hue=df['color'])
```

```python
sns.swarmplot(x=df['fare'],y=df['payment'],palette=['yellow','red'],hue=df['color'])
```

```python
sns.lineplot(x=df['total'],y=df['fare'])
```

```python
sns.lineplot(x=df['total'],y=df['fare'],hue=df['payment'],palette=['yellow','red'])
```

```python
sns.countplot(x=df['payment'])
```
