```python
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
df=read_csv('/content/pima-indians-diabetes.data.csv')
df.head()
```

```python
file_name='/content/pima-indians-diabetes.data.csv'
names=['preg','plas','pres','skin','test','mass','pedi','age','class']
df=read_csv(file_name,names=names)
df.head()
```

```python
array=df.values
x=array[:,0:8]
y=array[:,8]
```

```python
num_trees=100
seed=7
k_fold=KFold(n_splits=10,random_state=seed,shuffle=True)
model=RandomForestClassifier(n_estimators=num_trees,max_features=3)
k_fold
```

```python
result=cross_val_score(model,x,y,cv=k_fold)
print(result.mean())
```

```python
result=cross_val_score(model,x,y,cv=k_fold,scoring='accuracy')
print(result.mean())
```

```python
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
file_name="/content/pima-indians-diabetes.data.csv"
names=['preg','plas','pres','skin','test','mass','pedi','age','class']
df=read_csv(file_name,names=names)
df.head()
```

```python
array=df.values
array
```

```python
array=df.values
X=array[:,0:8]
Y=array[:,8]
```

```python
seed=7
k_fold=KFold(n_splits=10,random_state=seed,shuffle=True)
cart=DecisionTreeClassifier()
num_trees=100
model=BaggingClassifier(estimator=cart,n_estimators=num_trees,random_state=seed)
results=cross_val_score(model,X,Y,cv=k_fold)
print(results.mean())
```
